```python
#Google trained a famous model:
#GoogleNews-vectors-negative300
#Vocabulary: 3 million words
#Embedding size: 300

import gensim.downloader as api

#globals() contains all variables currently defined in the notebook.
if 'model' not in globals():
    print("Loading model...")
    model = api.load("word2vec-google-news-300")
else:
    print("Model already loaded.")

print(model.most_similar("india"))
```

```python
_model_cache = None

def get_word2vec():
    global _model_cache
    
    if _model_cache is None:
        print("Loading Word2Vec model...")
        _model_cache = api.load("word2vec-google-news-300")
    else:
        print("Using cached model")
    
    return _model_cache

model = get_word2vec()
print(model.most_similar("pakistan"))
```
