## 🔹 5. Transformer-Based Sentence Embeddings (BEST METHOD)

### 5.1 Sentence Transformers (Recommended)

Install: `pip install sentence-transformers`

✅ Understands meaning  
✅ Works for search  
✅ Industry standard  
✅ Best tradeoff of speed + accuracy

````python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = [
    "I love machine learning",
    "I like machine learning",
    "Deep learning is amazing"
]

embeddings = model.encode(sentences)
similarity_matrix = cosine_similarity(embeddings)
print(similarity_matrix)
````

## 🔹 6. OpenAI Embeddings (Production-Level Semantic Search)

If building enterprise-level search:

Install: `pip install openai`

✅ Best quality  
✅ Production-grade  
💲 Paid API

````python
from openai import OpenAI
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

client = OpenAI()

response = client.embeddings.create(
    model="text-embedding-3-small",
    input=["I love machine learning", "I like machine learning"]
)

embeddings = [e.embedding for e in response.data]
similarity = cosine_similarity([embeddings[0]], [embeddings[1]])
print(similarity)
````

## 🔹 9. BERT (Manual Implementation)

Install: `pip install transformers torch`

⚠ More complex  
⚠ Slower than sentence-transformers

````python
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

def embed(sentence):
    inputs = tokenizer(sentence, return_tensors="pt")
    outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).detach().numpy()

v1 = embed("I love machine learning")
v2 = embed("I like machine learning")

print(cosine_similarity(v1, v2))
````

## 🔥 Best Practical Code (Recommended)

If you want one production-ready solution:

````python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")

sentences = [
    "Reset my password",
    "I forgot my login password",
    "How to change email settings"
]

embeddings = model.encode(sentences)

def most_similar(query):
    query_vec = model.encode([query])
    similarities = np.dot(embeddings, query_vec.T).flatten()
    best_idx = np.argmax(similarities)
    return sentences[best_idx], similarities[best_idx]

print(most_similar("I can't log into my account"))
````
