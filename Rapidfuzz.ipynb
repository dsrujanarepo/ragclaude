```python
#1.2 Fuzzy Matching (RapidFuzz – Better than fuzzywuzzy)
#!pip install rapidfuzz
from rapidfuzz import fuzz
import pandas as pd

df1 = pd.DataFrame({"s1": ["I love ML", "Deep learning is fun"]})
df2 = pd.DataFrame({"s2": ["I like ML", "Deep learning rocks", "sujji"]})

# Create result columns first
df1["maxval"] = 0.0
df1["match"] = ""

fuzzy_matrix = pd.DataFrame(index=df1.index, columns=df2.index)

for i, text1 in df1["s1"].items():
    max_val=0
    for j, text2 in df2["s2"].items():
        score = fuzz.ratio(text1, text2)
        fuzzy_matrix.loc[i, j] = score
        if score > max_val:
            max_val = score
            best_match = text2
    df1.loc[i, "maxval"] = max_val
    df1.loc[i, "match"] = best_match
   
print(df1)
```

```python
from rapidfuzz import process
#🔥 This is MUCH faster.
#🧠 When To Use What

#Goal	Use
#Spell checking	fuzz.ratio
#Deduplication	fuzz.token_sort_ratio
#Sentence similarity	TF-IDF + cosine
#Meaning similarity	Sentence embeddings

fuzzy_matrix2 = process.cdist(
    df1["s1"],
    df2["s2"],
    scorer=fuzz.ratio
)

fuzzy_matrix2 = pd.DataFrame(
    fuzzy_matrix2,
    index=df1.index,
    columns=df2.index
)
print(fuzzy_matrix2)
```
