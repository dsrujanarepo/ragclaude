```python
#Example: Using BERT - sentiment analysis
#Using BERT as a finished product (a trained classifier)
#!pip install transformers
#!pip install torch

from transformers import pipeline

# Load sentiment analysis pipeline (uses a pretrained BERT-based model)
classifier = pipeline(
    task="sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

# Sentences to analyze
sentences = [
    "I love machine learning",
    "This movie was terrible"
]

# Run sentiment analysis
results = classifier(sentences)

# Print results
for sentence, result in zip(sentences, results):
    print(f"Sentence: {sentence}")
    print(f"Sentiment: {result['label']}, Confidence: {result['score']:.4f}")
    print()
```

```python
#Example: BERT is used for Embeddings
from transformers import BertTokenizer, BertModel
import torch
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased") #This model is NOT trained for sentiment.
model = BertModel.from_pretrained("bert-base-uncased")
text = "Machine learning is powerful"

inputs = tokenizer(text, return_tensors="pt")
print(inputs)
outputs = model(**inputs)
print(outputs)
embeddings = outputs.last_hidden_state
```

```python
#Because AutoTokenizer automatically selects the correct tokenizer based on the model.
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert-base-uncased-finetuned-sst-2-english"
)
text = "I love machine learning"
inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits
predicted_class = torch.argmax(logits)
print(predicted_class)
```

```python
#Example: Question Answering with BERT
import transformers
print(transformers.__version__)
from transformers import pipeline

qa = pipeline(
    task="question-answering",
    model="distilbert-base-cased-distilled-squad"
)

result = qa(
    question="Where is the Eiffel Tower?",
    context="The Eiffel Tower is located in Paris."
)

print(result)
```
