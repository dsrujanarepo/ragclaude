```python
#🔹 1. String-Based Similarity (Surface-Level Matching)
    #1.1 Levenshtein Distance (Edit Distance)
    #Measures number of edits required to convert one sentence to another.

#!pip install python-Levenshtein
import Levenshtein
import pandas as pd

df1 = pd.DataFrame({'s1': ["I love machine learning"]})
s1 = "I love machine learning"

df2 = pd.DataFrame({'s2': ["I like machine learning"]})
s2 = "I like machine learning"

distance = Levenshtein.distance(df1.loc[0, 's1'], df2.loc[0, 's2'])
distance1 = Levenshtein.distance(s1,s2)

ratio = Levenshtein.ratio(df1.loc[0, 's1'], df2.loc[0, 's2'])
ratio1 = Levenshtein.ratio(s1, s2)
```

```python
print("Edit Distance:", distance)
print("Edit Distance:", distance1)
print("Similarity Ratio:", ratio)
print("Similarity Ratio:", ratio1)

#✅ Good for typo detection
#❌ Not semantic
```
