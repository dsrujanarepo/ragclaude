```python
#🔹 4. Word Embeddings (Semantic Similarity)
#Now we move into meaning-based matching.
#4.1 Word2Vec (Average Word Vectors)
!pip install gensim
import gensim.downloader as api
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = api.load("word2vec-google-news-300")

def sentence_vector(sentence):
    words = sentence.split()
    vectors = [model[word] for word in words if word in model]
    return np.mean(vectors, axis=0)

v1 = sentence_vector("I love machine learning")
v2 = sentence_vector("I like machine learning")

similarity = cosine_similarity([v1], [v2])
print(similarity)

❌ Weak for long sentences
❌ Outdated compared to transformers
```
