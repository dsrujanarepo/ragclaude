```python
#2.2 TF-IDF + Cosine Similarity (Better)
#same like CountVectorizer but ..
#Term Frequency measures how often a word appears in a document
#Inverse Document Frequency reduces weight of common words across documents
#If a word appears in every document → IDF is small. If it appears in few documents → IDF is high
#TF-IDF=TF×IDF: Common words → low score : Rare but important words → high score

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

sentences = [
    "I love machine learning",
    "I like machine learning",
    "Deep learning is amazing"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)
print(X.toarray())

similarity_matrix = cosine_similarity(X)
print(similarity_matrix)

#✅ Good baseline
#✅ Works well for search systems
```

```python
#🔹 3. N-Gram Similarity. Controls single words or word combinations.
vectorizer = TfidfVectorizer(ngram_range=(1,2))
X = vectorizer.fit_transform(sentences)
print(X.toarray())

similarity_matrix = cosine_similarity(X)
print(similarity_matrix)

#🔹 3. max_features :: Limits vocabulary size.
#TfidfVectorizer(max_features=1000)

#🔹 4. min_df :: Ignore rare words.
#Keep words appearing in at least 2 documents.
#TfidfVectorizer(min_df=2)

#🔹 5. max_df :: Ignore too common words.
#Ignore words appearing in >80% of documents.
#TfidfVectorizer(max_df=0.8)

#🔹 6. sublinear_tf :: Uses log scaling:
#Helps when word frequency is very large.
#TfidfVectorizer(sublinear_tf=True)

#🔹 7. norm :: Normalization of vectors.
#Default: norm='l2'
#This makes vectors unit length. Useful for: Cosine similarity, SVM
```
