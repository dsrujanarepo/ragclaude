```python
#🔹 2. Bag of Words (Statistical Similarity)
#2.1 CountVectorizer + Cosine Similarity
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

sentences = [
    "I love machine learning",
    "I like machine learning",
    "Deep learning is amazing"
]

#create a super list of words. create a 3X n-words matrix where nth position has 1 or 0
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sentences)
print(X.toarray())

#CS of X.array gives a 3X3 matrix 
#CS(1,1), CS(1,2), CS(1,3)
#CS(2,1), CS(2,2), CS(2,3)
#CS(3,1), CS(3,2), CS(3,3)

similarity_matrix = cosine_similarity(X)
print(similarity_matrix)

#✅ Fast
#❌ Doesn't understand meaning
```
